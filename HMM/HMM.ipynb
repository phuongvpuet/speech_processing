```python
def get_data_by_id(student_id, df):
    """
    Dau vao la mssv va dataFrame (file excel), dau ra la 1 hang chua thong tin cua sinh vien do
    """
    for i in range(len(df)):
        if str(df.iloc[i]['Mã SV']).strip() == student_id.strip():
            return df.iloc[i]
    print("Not Found that student")
    return pd.core.series.Series()

def get_data_by_name(student_name, df):
    """
    Dau vao la ten sinh vien va dataFrame (file excel), dau ra la 1 hang chua thong tin cua sinh vien do
    """
    for i in range(len(df)):
        if str(df.iloc[i]['Họ và tên']).strip() == student_name.strip():
            return df.iloc[i]
    print("Not Found that student")
    return pd.core.series.Series()

def count_word(text):
    """
    Dem so tu xuat hien trong 1 van ban (text), tra ve 1 dictionary sap xep 
    theo thu tu pho bien nhat -> it xuat hien nhat
    """
    words = text.split()
    word_count = {}
    for word in words:
        if word in word_count:
            word_count[word] += 1
        else:
            word_count[word] = 1
    return sorted(word_count.items(), key=lambda kv: (kv[1], kv[0]), reverse=True)

def popular_word(*data, columns):
    """
    Dau vao la du lieu cua 1 hoac nhieu sinh vien tu function get_data va cac cot se duoc trich xuat
    dau ra la 1 dictionary dem so tu xuat hien trong du lieu,
    sap xep theo thu tu pho bien nhat -> it xuat hien nhat
    """
    all_text = ""
    for dt in data:
        for column in columns:
            aspect = dt[column].split('\n')[1:]
            for line in aspect:
                if line.strip()[-3:] != "wav":
                    line = line.translate(str.maketrans('', '', string.punctuation)).lower()
                    all_text += (line + " ")
    return count_word(all_text.strip())

def find_name_audio(words, data, columns):
    """
    Dau vao gom cac tu duoc chon (5 tu), du lieu cua 1 sinh vien, cac cot se duoc trich xuat
    Dau ra la ten cac file wav chua cac tu do 
    """
    audio_name = {}
    for word, _ in words:
        audio_name[word] = {}
        for column in columns:
            audio_name[word][column] = []
            aspect = data[column].split('\n')[1:]
            for idx, line in enumerate(aspect):
                if line.strip()[-3:] != "wav":
                    line = line.translate(str.maketrans('', '', string.punctuation)).lower()
                    if word in line:
                        audio_name[word][column].append(aspect[idx-1])
    return audio_name
```

```python
#Lay data cua dat va phuong
text_data_dat = get_data_by_name('Nguyễn Thành Đạt', df)
text_data_phuong = get_data_by_id('17020970', df)

#Seed 2020 
random.seed(2020)
arr = [x for x in range(0,100)]
#Chon 5 tu ngau nhien
random.shuffle(arr)
choose = arr[:5]
#--

words = popular_word(text_data_dat, text_data_phuong, columns=needed_columns)
choose_words = [words[i] for i in choose]
print("Chose Words: ",choose_words)

words_in_dat = find_name_audio(choose_words, text_data_dat, needed_columns)
words_in_phuong = find_name_audio(choose_words, text_data_phuong, needed_columns)
```
